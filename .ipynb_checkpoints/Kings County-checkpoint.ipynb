# Hypothesis Tests

```python
from statsmodels.formula.api import ols

lr_model = ols(formula='price~bedrooms+bathrooms+sqft_living+sqft_lot+waterfront+view+condition+grade+sqft_above+sqft_basement+yr_built+yr_renovated+zipcode+lat+long+sqft_living15+sqft_lot15+yr_built', data=df).fit()
lr_model.summary()
```

```python
lr_model = ols(formula='price~new_bathrooms', data=df).fit()
lr_model.summary()
```

```python
# Is there a relationship between no of bathrooms and sqft

# H0: mean(1)=mean(1.5)+mean(2)+mean(2.25)+mean(2.5)....
# H1: They are not all equal
# alpha = 0.05

# anova
stats, pvalue= st.f_oneway(df["new_bathrooms"], df["sqft_living"])
print("F statistics:",stats, ",pvalue:", pvalue)
```

```python
"""
Since the P value is 0.0 and less than alpha, the mean of each number of bathroom is not equal. 
"""
```

```python
lr_model = ols(formula='price~grade', data=df).fit()
lr_model.summary()
```

```python
#Is there a statistically significant difference in houses which are renovated and which are not renovated. 

# h0: mean(renovated) = mean(not_renovated)
# h0: mean(renovated) != mean(not_renovated)
# alpha=0.05

not_reno=df["yr_renovated"]==0
reno=df["yr_renovated"]!=0

# independent 2 sample T test
t_statictic, p_value = st.ttest_ind(not_reno, reno)
print("t_statictic:",t_statictic, ", pvalue:" , p_value)
```

```python
"""
Since the P value is 0.0, house being renovated or not is statistically significant.
"""
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features, target, random_state=40,test_size=0.2)
```

```python
selector = SelectKBest(f_regression, k=40)

selector.fit(X_train, y_train)
```

```python
selected_columns = X_train.columns[selector.get_support()]
removed_columns = X_train.columns[~selector.get_support()]
# selected_columns
```

```python
#instantiate a linear regression object
lm_kbest = linear_model.LinearRegression()

#fit the linear regression to the data
lm_kbest.fit(X_train[selected_columns], y_train)

y_train_kbest = lm_kbest.predict(X_train[selected_columns])


trainK_rmse = np.sqrt(metrics.mean_squared_error(y_train, y_train_kbest))


print('Training Root Mean Squared Error:' , trainK_rmse)

y_kbest = lm_kbest.predict(X_test[selected_columns])

testK_rmse = np.sqrt(metrics.mean_squared_error(y_test, y_kbest))

print('Testing Root Mean Squared Error:' , testK_rmse)
```

```python
print('Poly_2: ', test_rmse_2, '\n',
      "KBest:   ", testK_rmse,'\n')
```

```python
# recursive feature elimination
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features, target, random_state=10,test_size=0.2)
```

```python
ols = linear_model.LinearRegression()
# Create recursive feature eliminator that scores features by mean squared errors
selector = RFECV(estimator=ols, step=2, cv=7, scoring='neg_mean_squared_error', n_jobs=-1, verbose=1)

# Fit recursive feature eliminator 
selector.fit(X_train, y_train)
```

```python
ols = linear_model.LinearRegression()
```

```python
selected_rfe = X_train.columns[selector.support_]
removed_rfe = X_train.columns[~selector.support_]

len(selected_rfe)
```

```python
lm_rfe = linear_model.LinearRegression()

#fit the linear regression to the data
lm_rfe = lm_rfe.fit(X_train[selected_rfe], y_train)

y_rfe = lm_rfe.predict(X_train[selected_rfe])


trainRFE_rmse = np.sqrt(metrics.mean_squared_error(y_train, y_rfe))


print('Training Root Mean Squared Error:' , trainRFE_rmse)

y_pred_rfe = lm_rfe.predict(X_test[selected_rfe])

testRFE_rmse = np.sqrt(metrics.mean_squared_error(y_test, y_pred_rfe))

print('Testing Root Mean Squared Error:' , testRFE_rmse)
```

```python
print('Original: ', test_rmse_2, '\n',
      "KBest:   ", testK_rmse,'\n',
      "RFE:     ", testRFE_rmse)
```
